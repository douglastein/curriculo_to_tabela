```python
import os
from PyPDF2 import PdfReader

diretorio = 'C:/Users/dstei/OneDrive/Documentos/Projetos/Leitura_de_PDF/curriculos'

for nome_arquivo in os.listdir(diretorio):
    if nome_arquivo.endswith('.pdf'): # Verificando se o arquivo é pdf
        caminho_arquivo = os.path.join(diretorio, nome_arquivo) # Construindo o caminho para o arquivo
        leitor = PdfReader(caminho_arquivo)
        texto = ''
        
        for pagina in leitor.pages: # Percorrendo cada página do PDF
            texto += pagina.extract_text() # Extraindo o texto da página e adicionando à texto
            
        # Verificando a disponibilidade do texto
        print(texto)
```
